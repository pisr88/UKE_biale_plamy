8. Creating a DataFrame with coordinates 

```python
df_f = df[df['szerokosc_geo'].notnull() | df['dlugosc_geo'].notnull()]
```

9. Viewing individual elements on the map

- Indication of the town and street on the map

```python
city = input('Enter a city name to show on the map: ')
```

```python
street = input('Enter a street name to show on the map: ')
```

- Filtering the data to be displayed

```python
df_out = df_f[df_f['miejscowosc'].str.contains(city) & df_f['ulica'].str.contains(street)]
```

- Displaying items on the map

```python
from ipyleaflet import Map, Marker

# create map
center = (52.204793, 20)
m = Map(center=center, zoom=5)

# create markers
for (index, row) in df_out.iterrows():
    marker = Marker(location=[row.loc['szerokosc_geo'], row.loc['dlugosc_geo']], 
                    title=row.loc['miejscowosc'] + ' ' + row.loc['ulica'] + ' ' + row.loc['nr_porzadkowy'])
    m.add_layer(marker)
    
# basic_map
display(m)
```
